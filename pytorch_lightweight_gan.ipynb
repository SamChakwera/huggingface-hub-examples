## Train

```python
! lightweight_gan \
  --data ./jpg \
  --image-size 64 \
  --name flowers_gan_64 \
  --batch-size 16 \
  --gradient-accumulate-every 4 \
  --num-train-steps 10000 \
  --save_every 500 \
  --evaluate_every 500
```
